```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(num = 1,figsize=(14, 7))
#1 
#1xy   x
day = range(1,15)
#y
number_of_communities_CPM = [28,29,28,23,32,32,35,33,29,33,42,29,31,25]
number_of_communities_unweighted_Louvain = [8,8,7,8,8,9,8,9,8,8,9,9,9,9]
number_of_communities_weighted_Louvain = [11,12,10,12,13,12,13,12,11,12,11,13,12,10]
#2plot1x2y
plt.plot(day, number_of_communities_CPM , color="r", linestyle="-", marker="x", linewidth=1, label = 'CPM on unweighted graphs')
plt.plot(day, number_of_communities_unweighted_Louvain, color="b", linestyle="-", marker="^", linewidth=1, label = 'Louvain: on unweighted graphs')
plt.plot(day, number_of_communities_weighted_Louvain, color="g", linestyle="-", marker="*", linewidth=1, label = 'Louvain: on weighted graphs')

plt.axis([0, 15, 0, 60])

plt.xlabel("Day",fontsize = 14)
plt.ylabel("Number of communities",fontsize = 14)

plt.legend(loc='upper left', bbox_to_anchor=(0, 1),fontsize = 14)
plt.title(label = "Number of communities generated by different algorithms", pad = 10,fontsize = 15)

# 
for a, b in zip(day, number_of_communities_CPM):
    plt.text(a, b, b, ha='center', va='bottom', fontsize = 12 ) 
for a, b in zip(day, number_of_communities_unweighted_Louvain):
    plt.text(a, b, b, ha='center', va='top', fontsize = 12 )
for a, b in zip(day, number_of_communities_weighted_Louvain):
    plt.text(a, b, b, ha='center', va='bottom', fontsize = 12 )
#linesList = plt.plot(day, number_of_communities_CPM,  day, number_of_communities_unweighted_Louvain,  day, number_of_communities_weighted_Louvain )
#setp
#plt.setp(linesList, color='r', color='g', color='b')
plt.savefig('./number_communities.png')
plt.show()
```
